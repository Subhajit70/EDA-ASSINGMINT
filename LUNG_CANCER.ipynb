```python
x= df['LUNG_CANCER']
y=df['CHEST_PAIN']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train.values.reshape(-1, 1)) # Reshape x_train to a 2D array
x_test=sc.transform(x_test.values.reshape(-1, 1)) # Reshape x_test to a 2D array
```

```python
from sklearn.svm import SVC
classifier = SVC(random_state=0) # for non-linear model use this parametre kernel='rbf'
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
svc= SVC(C=100,kernel='rbf',degree=5,probability=True)
```
